# Extraccion de informacion del HTML

```python
import bs4
```

```python
soup = bs4.BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.title)
```

```python
print(soup.select('meta[name=description]'), '\n')
print(soup.select('meta[name=description]')[0]['content'])
```

```python
schools_links = soup.select('.SchoolsList-school')
schools = [school['href'] for school in schools_links]

for school in schools:
    print(school)
```

```python
base_url = 'https://www.infobae.com'
response = requests.get(base_url)

soup = bs4.BeautifulSoup(response.text, 'html.parser')
```

```python
news = soup.select('.cst_ctn')
for new in news:
    print(f'Title: {new["aria-label"]}\nURL: {base_url}{new["href"]}\n')
```
